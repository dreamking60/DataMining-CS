```python
md_Date = pd.to_datetime(md["date"])
md['date'] = md_Date
day = md['date'].dt.day
day_of_week = md['date'].dt.day_of_week
#day
```

```python
seaborn.distplot(day, kde=False, bins=31)
```

```python
seaborn.displot(day_of_week, kde=False, bins=[0,1,2,3,4,5,6,7])
```

```python
#Print out a lineplot
counts = day_of_week.value_counts().sort_index()
dayWeek = pd.DataFrame({'day_of_week': counts.index+1, 'counts': counts.values})
seaborn.lineplot(x='day_of_week', y='counts', data=dayWeek)
```
